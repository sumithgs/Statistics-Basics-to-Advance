# Problem
Suppose the IQ in a certain population is normally distributed with a mean of μ = 100 and a standard deviation of α = 15.

A researcher wants to know if a new drug affects IQ levels, so he recruits 20 patients to try it and records their IQ levels.

The following code shows how to perform a one sample Z-test in Python to determine if the new drug causes a significant ddifference in IQ levels:

```python
from statsmodels.stats.weightstats import ztest as ztest
data = [88, 92, 94, 94, 96, 97, 97, 97, 99, 99, 105, 109, 109, 109, 110, 112, 112, 113, 114, 115]

ztest(data, value=100)
# result Zscore or Ztest and P value - must be less than Significance value if not reject the null hypothesis
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import math
np.random.seed(6)
school_ages = stats.poisson.rvs(loc=18, mu=35, size=1500)
# poison distribution where we are considering loc = 18 start of our student age and mean mu = 35 and size of our population 1500.
classA_ages = stats.poisson.rvs(loc=18, mu=30, size=60)
```

```python
school_ages,classA_ages
```

```python
classA_ages.mean()
```

```python
sta,p_value = ttest_1samp(classA_ages, school_ages.mean())
```

```python
# accepted null hypothesis because 1.139027071016194e-13.
```

```python
if p_value < 0.05:
    print("Reject H0")
else:
    print("Accept H0")
```
