# Methods of identifying Outliers

## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Define Dataset

```python
dataset= [11,10,12,14,12,15,14,13,15,102,12,14,17,19,107, 10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]
```

### From the plot we can discover if the dataset has any outliers.

```python
plt.hist(dataset)
plt.show()
```

## Z score Computation

```python
outliers = []
def detect_outliers(data):
    threshold = 3 # setting threshold to 3 Standard Deviation
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
      z_score = (i-mean)/std
      if np.abs(z_score) > threshold:
        outliers.append(i)
    return outliers
```

```python
detect_outliers(dataset)
```

#### Note: We can calculate a Z-score for any dataset, but using it to decide if a point is an outlier is only strictly appropriate for normal (or very close to normal) data. For non-normal data, it can be misleading.

## IQR
### In IQR we find lower and higher fence.

Steps Involved:
1. Sort the dataset
2. Caculate Q1 and Q3 (25th Percentile and 75th Percentile)
3. IQR = Q3 - Q1
4. Lower Fence (Q1 - 1.5(IQR))
5. Higher Fence (Q3 + 1.5(IQR))

```python
dataset = sorted(dataset)
dataset
```

```python
q1,q3 = np.percentile(dataset,[25,75])
q1, q3
```

```python
lFence = q1 - (1.5*(q3-q1))
hFence = q3 + (1.5*(q3-q1))
lFence, hFence
```

#### The IQR method is a general-purpose, robust technique for outlier detection that is suitable for any distribution type.
